```python
def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

```python
pd.set_option('float_format', '{:f}'.format)
```

```python
tn.worldwide_gross = tn.worldwide_gross.apply(clean_currency).astype(np.int64)
```

```python
tn.sort_values('worldwide_gross', ascending=False)
```

```python
tn.domestic_gross = tn.domestic_gross.apply(clean_currency).astype(np.int64)
tn.sort_values('domestic_gross', ascending=False)
```

```python
tn.production_budget = tn.production_budget.apply(clean_currency).astype(np.int64)
tn.sort_values('production_budget', ascending=False)
```

```python
tn.total_profit = (tn.worldwide_gross + tn.domestic_gross) - tn.production_budget
```
